# InetCampus operation - Email servers budget study #

------------------------------------------------

## 1.- Analytic Approach ##

The servers could be classified in incoming mails servers and outbound mails servers.

The analysis should also take in consideration that the company uses emails to establish communications with the customers and internal emails between employees should also be considered.

__InetCampus__ stores the emails and their characteristics for five years, passed this time the information is progressively deleted. _In the moment of the acquisition __InetCampus__ is trying to rescue information of five years more recovering erased data from the backup disks._

Based in the nature of the information we are dealing with, the study should be founded over values associated to the use of emails, as the number of replies, the average of messages sent per person, etc, all this data generated during a certain period of time, bearing in mind that the temporal horizon of the analysis should never go further than five years.

The integration of __InetCampus__ as part of a bigger company implies an adaptation of its policies and rules in the field of communications and the values valid during all its life may somehow change, this should be considered during the construction of the models.

## 2.- Data Requirements ##

Information in the study includes:

1. Data about emails sent and received by customers from __InetCampus__.

2. Data about emails sent and received by employees from __InetCampus__.

3. Data about emails sent and received by employees from the parent company.

Data about customers and employees should be in this case normalized and treated homogeneously, however a column should differentiate their roles and the company source of the information.

The rules and limitations in the use of emails in the parent company should be collected in order to obtain the future values associated to the employees transferred in the acquisition of __InetCampus__.

## 3.- Data Collection ##

The sources of the data needed for the study are:

1. Statistical data collected from __InetCampus__ with a maximum period of time of five years.

2. Politics and rules in the use of messages applicable to employees of __InetCampus__.

3. Register of issues caused by wrong destination that were detected by the servers of ___InetCampus___.

4. History of attacks with denial of service suffered monthly by the servers of ___InetCampus___ that could deviate the statistics.

5. Politics and rules in the use of messages applicable to employees of the parent company.

6. _By now and pending the recovering of erased data from the backup disks, the processing of statistical old data with more than five years rescued by __InetCampus__ has been deferred._


## 4.- Data Understanding and Preparation ##

The data collected should be analyzed searching for missing, duplicated, incorrectly formatted and invalid values

All the information should be aggregated creating one transactional record per thread, including information about the subject, the number of emails sent, the interlocutors, the number of internet providers that participate in the thread, the role of the initial sender and the amount of bytes of the thread. 

A thread is determined by its initial interlocutors and its initial subject (not considering prefixes added as "Re", "Fwd", etc), if one of this attributes change a new transactional record should be created.

_The data referring to emails generated by the company in the exercise of its functions should be incorporated to the model that should be considered an iterative process that provides information about future requirements in the servers infrastructure._

## 5.- Modeling and Evaluation ##

The models created are based in the number of the emails analyzed in each thread in order to calculate the disk space needed to store all the text.

| Model  | # Emails considered | Cost | Eficiency  |
|---|---|---|---|
| 1  | 1 | 1 | 20% |   
| 2  | 5  | 5  | 35% |   
| __3__  | __7__  | __6__  | __85%__ |   
| 4  | 15 | 9.2 | 98.5% |   

The first model only obtains information from the initial email. In most of the cases a thread includes at least two messages. The process of calculating the number of bytes of the thread is in this case fast and the disk space used is low but the results in this model are not accurate.

In the other hand, the fourth model analyzes a total of 15 emails in average which gives a precise result but the calculus is expensive in time and disk space.

__Finally and considering the pros and cons, the third model is selected because it presents the best balance between results accuracy and time and disk space cost.__
